```python
import torch

# Load pretrained YOLOv5s model (or custom model with path='best.pt')
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Optional: set to eval mode
model.eval()

# Load an image (you can upload your own image first)
img_path = '/content/ChatGPT Image Apr 9, 2025, 10_33_58 AM.png'  # Example image
results = model(img_path)

# Show image with detections (optional)
results.show()

# Extract detected object names
detected_items = results.pandas().xyxy[0]['name'].tolist()
unique_items = list(set(detected_items))

# Print results
print("Detected Items:", unique_items)





```

```python
# Sample recipe database
recipe_db = {
    "Tomato Pasta": ["tomato", "pasta", "onion"],
    "Omelette": ["egg", "onion", "salt"],
    "Fruit Salad": ["apple", "banana", "milk"],
    "Simple Salad": ["tomato", "cucumber", "salt"]
}

def suggest_recipes(inventory):
    suggestions = []
    for recipe, ingredients in recipe_db.items():
        matched = [item for item in ingredients if item in inventory]
        if matched:  # Suggest if at least one ingredient is present
            suggestions.append({
                "recipe": recipe,
                "matched": matched,
                "total": len(ingredients),
                "match_percent": int(len(matched) / len(ingredients) * 100)
            })
    # Sort by best match
    suggestions.sort(key=lambda x: x["match_percent"], reverse=True)
    return suggestions

# Example inventory (for testing)
inventory = ["tomato", "salt", "apple"]

# Get recipe suggestions
suggested = suggest_recipes(inventory)

# Display results
print("Suggested Recipes:")
for item in suggested:
    print(f"- {item['recipe']} ({item['match_percent']}% match: {', '.join(item['matched'])})")
```

```python
from IPython.display import display
import ipywidgets as widgets

# Display Detected Items
items_label = widgets.Label(value="Detected Items: " + ", ".join(unique_items))
display(items_label)

# Display Suggested Recipes
if suggested:
    recipes_label = widgets.Label(value="Suggested Recipes:")
    recipe_list = widgets.HTML(value="<ul>" + "".join(f"<li>{r['recipe']} ({r['match_percent']}%)</li>" for r in suggested) + "</ul>")
    display(recipes_label, recipe_list)
else:
    display(widgets.Label(value="No matching recipes found."))

```
